```
annotations = pd.read_csv('/content/drive/My Drive/Data/Head Counts/bbox_train.csv',
                          usecols = ['Name',
                                                             'label',
                                                             'xmin',
                                                             'ymin',
                                                             'xmax',
                                                             'ymax',
                                                             'width',
                                                             'height'])
annotations.head()
```

```
annotations['xmin'] = annotations['xmin']/annotations['width']
annotations['xmax'] = annotations['xmax']/annotations['width']
annotations['ymin'] = annotations['ymin']/annotations['height']
annotations['ymax'] = annotations['ymax']/annotations['height']

annotations['center x'] = (annotations['xmin'] + annotations['xmax'])/2
annotations['center y'] = (annotations['ymin'] + annotations['ymax'])/2

annotations.head()
```

```
import pandas as pd
import numpy as np
import cv2
import glob
import time
import matplotlib.pyplot as plt
```

```
test = glob.glob('/content/drive/My Drive/Data/Head Counts/test/*')
print(len(test))
test[0][46:]
```

```
image = []
string = []
total_obj = []

for i in test_images:
    
    img = '/content/drive/My Drive/Data/Head Counts/test' + '/' +i
   
    
    image_BGR = cv2.imread(img)
    image.append(img[46:])
    

    h,w = image_BGR.shape[:2]


    blob = cv2.dnn.blobFromImage(image_BGR, 1/255, (416,416), swapRB =True, crop= False)

    blob_to_show = blob[0,:,:,:].transpose(1,2,0)

    with open('/content/drive/My Drive/Data/Head Counts/classes.names') as f:
        labels = [line.strip() for line in f]

   
    network = cv2.dnn.readNetFromDarknet('/content/drive/My Drive/Data/Head Counts/yolov3_custom_train.cfg',
                                         '/content/drive/My Drive/Data/Head Counts/yolov3_custom_train_last_10.weights')

    
    layers_names_all = network.getLayerNames()
    

    layers_names_output = [layers_names_all[i[0] - 1] for i in network.getUnconnectedOutLayers()]

  

    probability_minimum = 0.1

    threshold = 0.5

    colours = np.random.randint(0,255, size = (len(labels) ,3), dtype = 'uint8')


    network.setInput(blob)
    start = time.time()    
    output_from_network = network.forward(layers_names_output)
    end = time.time()


    bounding_boxes = []
    confidences = []
    class_numbers = []

    for result in output_from_network:
        for detected_objects in result:
            scores = detected_objects[5:]
            
            class_current = np.argmax(scores)
            confidence_current = scores[class_current]
        
        
            if  confidence_current > probability_minimum:
                box_current = detected_objects[0:4] * np.array([w,h,w,h])
                x_center, y_center, box_width, box_height = box_current
                x_min = int(x_center - (box_width/2))
                y_min = int(y_center - (box_height/2))
                
                bounding_boxes.append([x_min, y_min, int(box_width), int(box_height)])
                confidences.append(float(confidence_current))
                class_numbers.append(class_current)
            
        

        
    result = cv2.dnn.NMSBoxes(bounding_boxes, confidences, probability_minimum, threshold)

    pred_strings = []

    counter = 1
    if len(result)>0:
    
        for i in result.flatten():
            
            counter +=1
        
            x_min, y_min = bounding_boxes[i][0], bounding_boxes[i][1]
        
            box_width, box_height = bounding_boxes[i][2], bounding_boxes[i][3]
        
            colour_box_current = colours[class_numbers[i]].tolist()
            
        
            pred_strings.append("{0:.4f} {1} {2} {3} {4}".format(confidences[i], x_min, y_min, box_width, box_height))
            
        
            cv2.rectangle(image_BGR, (x_min,y_min),
                         (x_min + box_width, y_min+box_height),
                         colour_box_current,2)
        
            text_box_current = '{}: {:.4f}'.format(labels[int(class_numbers[i])],
                                                  confidences[i])
        
            cv2.putText(image_BGR, text_box_current, (x_min, y_min-5),
                       cv2.FONT_HERSHEY_COMPLEX, 0.7, colour_box_current, 2)
    
    pred_strings = str(pred_strings).replace(',','')
    pred_strings = str(pred_strings).replace('[','')
    pred_strings = str(pred_strings).replace(']','')
    pred_strings = str(pred_strings).replace("'",'')
    string.append(pred_strings)
    
    total_obj.append(counter-1)

    print()
    print('Image:',img[46:])
    print('Total objects been detected: ',len(bounding_boxes))
    print('Number of objects left after non-maximum supprssion: ',counter-1)
```
